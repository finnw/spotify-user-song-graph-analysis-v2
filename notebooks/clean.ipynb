This notebook contains code to clean up intermediate forms of the dataset, in case they are invalidated by changes to other code in particular `pipeline.py`

Import the essential libraries that we need

```python
import os
```

Set the working directory to the project root to simplify access to the data and other Python modules

```python
current_dir = os.getcwd()
project_directory, notebooks_subdirectory = os.path.split(current_dir)
if notebooks_subdirectory == "notebooks":
    os.chdir(project_directory)
    print(f"Changed working directory to project root: {project_directory}")
else:
    project_directory = current_dir
# Confirm the new current directory
os.getcwd()
```

Also add our local modules to the import path

```python
if project_directory not in os.sys.path:
    os.sys.path.append(project_directory)
```

```python
full = True
basenames = [
    "filtered_listens.csv", "filtered_songs.csv", "filtered_users.csv",
    "threshold_stats.csv", "user_embeddings.csv", "song_embeddings.csv",
    "user_classes.csv", "song_classes.csv"
]
if full:
    basenames.extend(["stats.csv", "listens.csv", "songs.csv", "users.csv"])
for basename in basenames:
    file_path = os.path.join(project_directory, "data", basename)
    if os.path.exists(file_path):
        print(f"Removing existing file: {file_path}")
        os.remove(file_path)
```
